```python
# Importação de bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap

# Configurações gerais com melhorias visuais
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (14, 8)
plt.rcParams['font.size'] = 12
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12

# Definindo uma paleta de cores personalizada mais agradável
cores = ['#4C72B0', '#55A868', '#C44E52', '#8172B3', '#CCB974', '#64B5CD']

# Carregamento da base de dados
data = pd.read_csv("assets/data/pemob_metropolitana_2024.csv")

# Visualização inicial dos dados
print("Primeiras linhas da base de dados:")
display(data.head())

# Estrutura da base de dados
print("\nInformações estruturais da base de dados:")
data.info()

# Estatísticas descritivas com formatação melhorada
print("\nEstatísticas descritivas:")
desc_stats = data.describe().round(2)
display(desc_stats)

# Verificação de valores ausentes (melhorado)
missing_values = data.isnull().sum()
missing_percentage = (missing_values / len(data)) * 100
missing_df = pd.DataFrame({
    "Coluna": data.columns,
    "Valores Ausentes": missing_values,
    "Porcentagem (%)": missing_percentage.round(2)
})
missing_df = missing_df.sort_values(by="Porcentagem (%)", ascending=False)

# Mostrando apenas colunas com valores ausentes
missing_df_filtered = missing_df[missing_df["Valores Ausentes"] > 0]
print("\nColunas com valores ausentes:")
display(missing_df_filtered)

# Visualização gráfica melhorada dos valores ausentes
plt.figure(figsize=(12, 8))
missing_top10 = missing_df_filtered.head(10)
bars = sns.barplot(x=missing_top10["Porcentagem (%)"], y=missing_top10["Coluna"],
                  palette=sns.color_palette("viridis", n_colors=len(missing_top10)))

# Adicionando valores nas barras
for i, p in enumerate(bars.patches):
    width = p.get_width()
    plt.text(width + 0.5, p.get_y() + p.get_height()/2,
             f'{width:.2f}%', ha='left', va='center', fontweight='bold')

plt.title("Top 10 Colunas com Mais Valores Ausentes", fontweight='bold', fontsize=18, pad=20)
plt.xlabel("Porcentagem de Valores Ausentes (%)", fontweight='bold')
plt.ylabel("Colunas", fontweight='bold')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Distribuição das colunas numéricas com layout e cores melhoradas
numerical_cols = data.select_dtypes(include=["float64", "int64"]).columns

if len(numerical_cols) > 0:
    n_cols = 3
    n_rows = (len(numerical_cols) - 1) // n_cols + 1

    fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, n_rows * 4))
    axes = axes.flatten() if hasattr(axes, 'flatten') else [axes]

    for i, col in enumerate(numerical_cols):
        if i < len(axes):
            sns.histplot(data[col].dropna(), kde=True, ax=axes[i],
                         color=cores[i % len(cores)],
                         edgecolor='white', alpha=0.7, bins=15)

            axes[i].set_title(f'Distribuição: {col}', fontweight='bold')
            axes[i].set_xlabel(col)
            axes[i].set_ylabel('Frequência')
            axes[i].grid(alpha=0.3, linestyle='--')

            # Adicionando estatísticas básicas no gráfico
            mean_val = data[col].mean()
            median_val = data[col].median()

            if not pd.isna(mean_val) and not pd.isna(median_val):
                axes[i].axvline(mean_val, color='red', linestyle='--', alpha=0.8,
                               label=f'Média: {mean_val:.2f}')
                axes[i].axvline(median_val, color='green', linestyle='-', alpha=0.8,
                               label=f'Mediana: {median_val:.2f}')
                axes[i].legend(frameon=True, facecolor='white', framealpha=0.9)

    # Removendo subplots vazios
    for j in range(i+1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.subplots_adjust(top=0.92, hspace=0.4)
    plt.suptitle("Distribuição das Variáveis Numéricas", fontsize=20, fontweight='bold')
    plt.show()

# Mapa de calor para correlação com design melhorado
if len(numerical_cols) > 1:
    plt.figure(figsize=(14, 12))

    # Criando uma máscara para mostrar apenas o triângulo inferior (evita redundância)
    corr_matrix = data[numerical_cols].corr().round(2)
    mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

    # Criando um colormap personalizado mais atraente
    cmap = LinearSegmentedColormap.from_list('custom_cmap', ['#053061', '#f7f7f7', '#67001f'])

    # Gerando o mapa de calor
    sns.heatmap(corr_matrix, annot=True, cmap=cmap,
                fmt=".2f", linewidths=0.8, mask=mask,
                cbar_kws={"shrink": .8, "label": "Coeficiente de Correlação"},
                annot_kws={"size": 12})

    plt.title("Mapa de Calor - Correlação Entre Variáveis Numéricas",
              fontweight='bold', fontsize=18, pad=20)
    plt.xticks(rotation=45, ha='right', fontsize=12)
    plt.yticks(fontsize=12)
    plt.tight_layout()
    plt.show()

# Análise das categorias dominantes com gráficos aprimorados
categorical_cols = data.select_dtypes(include=["object"]).columns

if len(categorical_cols) > 0:
    for col in categorical_cols[:5]:  # Primeiras 5 colunas categóricas
        print(f"\n\nDistribuição da coluna '{col}':")
        value_counts = data[col].value_counts().head(10)

        # Exibindo tabela de frequências
        freq_df = pd.DataFrame({
            f"{col}": value_counts.index,
            "Contagem": value_counts.values,
            "Percentual (%)": (value_counts.values / value_counts.sum() * 100).round(2)
        })
        display(freq_df)

        # Criando gráfico de barras com visual aprimorado
        plt.figure(figsize=(14, 6))

        # Criando gráfico de barras horizontal para melhor visualização de categorias com nomes longos
        bars = sns.barplot(y=value_counts.index, x=value_counts.values,
                          palette=sns.color_palette(cores, n_colors=len(value_counts)))

        # Adicionando rótulos nas barras
        for i, p in enumerate(bars.patches):
            width = p.get_width()
            plt.text(width + 0.5, p.get_y() + p.get_height()/2,
                     f'{width:,.0f} ({(width/value_counts.sum()*100):.1f}%)',
                     ha='left', va='center')

        plt.title(f'Top 10 Categorias - {col}', fontweight='bold', fontsize=18)
        plt.xlabel('Contagem', fontweight='bold')
        plt.ylabel(col, fontweight='bold')
        plt.grid(axis='x', linestyle='--', alpha=0.7)
        plt.tight_layout()
        plt.show()

        # Adiciona um gráfico de pizza para visualização alternativa (apenas para poucas categorias)
        if len(value_counts) <= 7:
            plt.figure(figsize=(10, 10))
            plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%',
                   colors=cores[:len(value_counts)], explode=[0.05]*len(value_counts),
                   textprops={'fontsize': 12}, shadow=True)
            plt.title(f'Distribuição de {col} (%)', fontweight='bold', fontsize=18)
            plt.axis('equal')
            plt.tight_layout()
            plt.show()

# Salvando o DataFrame limpo para análises futuras
# data.to_csv('/content/pemob_metropolitana_2024_processado.csv', index=False)

print("\n✅ Análise exploratória concluída com sucesso!")
```
